```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly
```

```python
df = pd.read_csv('../data/master.csv', parse_dates=['visit_date'])
```

```python
df.head()
```

```python
# SPLIT, APPLY, COMBINE
# Split the dataset up into 829 rows, 
#    applying the sum() function to each feature, 
#    and then combine the features into a table

# in my example i drop the latitude and longitude columns

df.groupby('id').sum().drop(columns=['latitude', 'longitude'])
```

```python
df.groupby('holiday').sum().drop(columns=['latitude', 'longitude'])
```

```python
df.groupby(['day_of_week', 'holiday'])\
    .mean().drop(columns=['latitude', 'longitude'])\
    .sort_index(level=1, ascending=False)\
    .reset_index()
```

```python
def is_missing(series):
    return pd.isnull(series).sum()
```

```python
df.groupby(['day_of_week', 'holiday']).agg([
    'count', np.mean, 'sum', 'std', is_missing
]).drop(columns=['latitude', 'longitude'])
```
